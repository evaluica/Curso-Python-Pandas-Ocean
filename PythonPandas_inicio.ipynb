```
# DataFrame

# Representa uma matriz/dataset, que são tabelas

# Matriz no python é uma lista com várias listas dentro
# Um DataFrame possui várias Series dentro

# Criação do Dicionário

dicionario = {
    'ESTADO': ['Amazonas', 'Amapá', 'Acre'],
    'SIGLA': ['AM', 'AP', 'AC'],
    'CAPITAL': ['Manaus', 'Macapá', 'Rio Branco']
}

# Acessando as informações do dicionário utilizando apenas Python

print(type(dicionario))
print(dicionario)

capitais = dicionario['CAPITAL']

print(type(capitais))
print(capitais)

capitalAmazonas = capitais[0]

print(type(capitalAmazonas))
print(capitalAmazonas)

 # Utilizar o dicionário para criar o DataFrame

dt = pd.DataFrame(dicionario)

# Visualizando o DataFrame

print(type(dt))
print(dt)
```
